```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
import random
```

```python
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False#对列表的格式进行修改
x = np.arange(0.05, 5, 0.02)
#定义一个常函数
y1 = [6 for i in x ]
plt.plot(x, y1, linewidth = 2, label = '常函数：y = 5')
#定义一个一次函数
y2 = [2*i + 1 for i in x]
plt.plot(x, y2, linewidth = 2, label = '一次函数：y = 2x + 1')
#定义一个二次函数
y3 = [1.5*i*i - 3*i - 2 for i in x]
plt.plot(x, y3, linewidth = 2, label = '二次函数: y = 1.5$x^2$ - 3x -2 ')
#定义一个对数函数
y4 = [math.log(i, 2) for i in x]
plt.plot(x, y4, linewidth = 2, label = '对数函数：y = log2(x)')
#定义一个指数函数
y5 = [math.pow(2, i) for i in x]
plt.plot(x, y5,linewidth = 2, label = '指数函数：y =$2^x$')
#幂函数
y6 = [math.pow(i, 2) for i in x]
plt.plot(x, y6, linewidth = 2, label = '幂函数: y = i^2')
plt.legend(loc= 'upper left')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0.05, 10, 0.02)
y1 = [math.log(i, 0.5) for i in x]
plt.plot(x, y1,linewidth= 2, color = 'y',label= 'log0.5(x)')
y2 = [math.log(i, math.e) for i in x]
plt.plot(x, y2, linewidth = 2, color = 'b', label = 'ln(x)')
y3 = [math.log(i, 5) for i in x]
plt.plot(x, y3, linewidth = 2, color = 'g', label = 'log5(x)')
y4 = [math.log(i, 10) for i in x] 
plt.plot(x, y4, linewidth = 2, color = 'r', label = 'log10(x)')
plt.plot([1,1],[-3,4],'-',color = '#999999',linewidth = 2)
plt.legend = loc = ('lower right')
plt.grid = True
plt.show()
```

```python
#大数定律解释
#确保中文能够正常显示
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False
#给定一个随机数的种子
random.seed(25)
#定义一个函数来得到1-9的随机数、共经历n次
def get_random_number(n):
    return[random.randint(1,9) for i in range(n)]
#定义主函数来显示大数定律
if __name__ == '__main__':
    number = 10000
    #想让这次实验经过多少次
    x = [i for i in range(number+1) if i != 0]
    #调用函数产生number的（1-9）的随机数
    total_randomint = get_random_number(number)
    #设定y轴，其中Y轴就是这些随机数的均值,由于是要求k+1项的随机数的均值，所以考虑到应该使用切片的方式
    y = [np.mean(total_randomint[0 : i+1]) for i in range(number)]
    
    plt.plot(x, y, 'b-', linewidth = 2, color = 'y')
    plt.xlim(0,number)
    plt.grid == True
    plt.show()
    
        
```
